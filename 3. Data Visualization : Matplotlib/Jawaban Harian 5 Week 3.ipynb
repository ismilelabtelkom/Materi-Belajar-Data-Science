```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Quiz 1 Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
data = df.groupby('Genre').agg({'NA_Sales' : 'mean', 'EU_Sales' : 'mean', 'JP_Sales' : 'mean', 'Other_Sales' : 'mean'})
display(data)
```

<h2>Perbandingan Kuantitatif Barplot : Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
data    = data.T
x       = np.arange(1,8,2)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=[20,10])

for i, col in enumerate(data.columns):
    ax.bar(x + i*0.1, data[col], width=0.1, label=col)


plt.xticks(x+0.5,list(data.index))
ax.set_title('Mean Sales Video Games by Genre', fontweight = 'bold')
ax.set_ylabel('Mean Sales')
ax.set_xlabel('Region Sales')
plt.legend()
plt.show()
```

# 2. Insight yang didapatkan:
