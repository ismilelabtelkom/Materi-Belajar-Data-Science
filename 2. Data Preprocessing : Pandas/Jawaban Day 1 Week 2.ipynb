<h2>Quiz 3 : DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

* series: merupakan objek 1 dimensi dan biasa dibentuk menggunakan pandas, list: objek 1 dimensi yang dapat berisi berbagai jenis tipe data didalamnya  dict: objek 1 dimensi yang terdiri dari key (kunci) dan value (nilai) dan dapat berisi berbagai jenis tipe data

* dataframe suatu objek 2 dimensi yang mmemiliki bentuk seperti tabel sedangkan series merupakan objek 1 dimensi

```python
# code here
nama = pd.Series(['Sung jun woo','Siti','Jojon','BUdi'],index=['a','b','c','d'])
print(nama)
```
